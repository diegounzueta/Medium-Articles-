```python
df = pd.read_csv("./data/Properties For Sale in London _ Rightmove_all_property_areas.csv", index_col = 0)
```

```python
df.head()
```

```python
def pre_processing(df):
    
    df = df[["Title", "Price", "Type", "Bedrooms", "Bathrooms","SQ M"]]
    
    #BEDROOMS BATHROOMS
    for i in ["Bedrooms","Bathrooms"]:
        df[i].replace("NaN", np.nan).astype(float)
    
    
    #PRICE
    p = []
    for i in df["Price"]:    
        if ("POA" in i):
            p.append(np.nan)
        else:
            p.append(int(i[1:].replace(",","")))
    df["Price"] = p


    cords = []
    for address in df["Title"]:
        print(address)
        location = gmaps_client.geocode(address)
        try:
            cords.append([location[0]["geometry"]["location"]["lat"], location[0]["geometry"]["location"]["lng"]])
        except:
            cords.append([np.nan,np.nan])
    cords = np.array(cords)

    df["latitude"] = cords[:, 0]
    df["longitude"] = cords[:, 1]
    return(df)
```

```python
df = pre_processing(df)
```

```python
def plot_london(lat, lng, zoom=11, map_type='roadmap', alpha = 0.3):
    
    hover = HoverTool(
    tooltips = [
        ('Address', '@Title'),
        ('Price', '@Price'), 
        ('Sq Meters', '@sq_meters'),
        ('Type', '@Type')
    ]
    )
    mapper = linear_cmap('Price', color, 0, df_plot["Price"].max())
    
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    
    bokeh_width, bokeh_height = 1000,700
    p = gmap(api_key, gmap_options, 
             width=bokeh_width, height=bokeh_height, tools=[hover, 'reset', 'wheel_zoom', 'pan'])
    
    
    source = ColumnDataSource(df_plot)
    center = p.circle("longitude", "latitude", color = mapper, size="area_scaled", alpha=0.8, source = source )
    color_bar = ColorBar(color_mapper=mapper['transform'], 
                         location=(0,0))
    p.add_layout(color_bar, 'right')
    show(p)
    return p
```
