```python
s = pd.DataFrame(sensitive_analysis)
```

```python
stocks_mention = {}
for index, row in stocks.iterrows():
    if row["names"] not in stocks_mention:
        stocks_mention[row["names"]] = {}
        stocks_mention[row["names"]]["texts"] = []
        stocks_mention[row["names"]]["sentiment"] = []
        stocks_mention[row["names"]]["confidence"] = []
        
    for _, i in s.iterrows():
        if i["start"] < row["start"] and i["end"] > row["end"]:
            stocks_mention[row["names"]]["texts"].append(i["text"])
            stocks_mention[row["names"]]["sentiment"].append(i["sentiment"])
            stocks_mention[row["names"]]["confidence"].append(i["confidence"])
```
